### Bias-Variance Trade-Off

#### Chap 2-Question 3-Sub (A)

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Flexibility axis

```python
flexibility = np.linspace(0.1, 10, 100)
```

#### Curves (theoretical shapes)

```python
bias_squared = 1 / flexibility
variance = flexibility / 10
irreducible_error = np.ones_like(flexibility) * 0.2


training_error = bias_squared * 0.6
test_error = bias_squared + variance + irreducible_error
```

#### Plot

```python
plt.figure(figsize=(8, 6))
plt.plot(flexibility, bias_squared, label="Squared Bias")
plt.plot(flexibility, variance, label="Variance")
plt.plot(flexibility, training_error, label="Training Error")
plt.plot(flexibility, test_error, label="Test Error")
plt.plot(flexibility, irreducible_error, linestyle="--", label="Bayes (Irreducible) Error")


plt.xlabel("Model Flexibility")
plt.ylabel("Error")
plt.title("Bias-Variance Trade-off")
plt.legend()
plt.show()
```

#### Sub (B)

## Explanation of Why Each Curve Has Its Shape

We explain the behavior of each curve as model flexibility increases.

---

### 1 Squared Bias  *Decreases*

Inflexible models (for example, linear regression) tend to oversimplify real-world relationships.

As model flexibility increases, the model becomes better able to approximate the true underlying function \( f \).  
As a result, the systematic error caused by incorrect model assumptions is reduced.

Therefore, the **squared bias decreases** as flexibility increases.

**Example:**  
Linear regression fitting a highly nonlinear relationship results in high bias.

---

### 2 Variance  *Increases*

Flexible models are highly sensitive to the training data.

Small changes or noise in the training dataset can lead to large changes in the fitted model.  
This sensitivity causes predictions to vary significantly across different training samples.

Therefore, the **variance increases** as model flexibility increases.

**Example:**  
A high-degree polynomial that closely follows every data point.

---

### 3 Training Error  *Always Decreases*

As flexibility increases, a model can fit the training data more closely.

Highly flexible models may even memorize the training data, a phenomenon known as **overfitting**.  
Because the model is optimized to minimize error on the training set, the training error never increases.

**Key exam statement:**  
> *Training error is a non-increasing function of model flexibility.*

---

### 4 Test Error  *U-shaped*

From Equation (2.7):

\[
\text{Test MSE} = \text{Bias}^2 + \text{Variance} + \text{Irreducible Error}
\]

- Initially, increasing flexibility causes bias to decrease faster than variance increases, so test error decreases.
- After a certain point, variance increases rapidly and dominates the bias reduction, causing test error to increase.

This produces a **U-shaped curve**, known as the **biasvariance trade-off**.

---

### 5 Bayes (Irreducible) Error  *Flat Line*

The Bayes (irreducible) error arises from inherent noise \( \varepsilon \) in the data.

This error cannot be reduced by any statistical learning method, regardless of model flexibility.  
It represents the **lower bound** on the achievable test error.

**Important quote:**  
> *No model can do better than the Bayes error.*
