## (a) Function Power()

```python
def Power():
    print(2**3)

Power()
```

## (b) Function Power2(x, a)

```python
def Power2(x, a):
    print(x**a)

Power2(3, 8)   # Output: 6561
```

## (c) Using Power2()

```python
Power2(10, 3)
Power2(8, 17)
Power2(13, 13)
```

## (d) Function Power3(x, a) (returns value)

```python
def Power3(x, a):
    result = x**a
    return result

print(Power3(3, 4))
```

## (e) Plot of ùëì(ùë•)=ùë•^2

```python
import matplotlib.pyplot as plt

x = range(1, 11)
y = [Power3(i, 2) for i in x]

plt.plot(x, y, marker='o')
plt.xlabel("x")
plt.ylabel("x^2")
plt.title("Plot of f(x) = x^2")
plt.yscale("log")   # optional log scale
plt.show()
```

## (f) Function PlotPower(x, a)

```python
import matplotlib.pyplot as plt

def PlotPower(x, a):
    y = [i**a for i in x]
    plt.plot(x, y, marker='o')
    plt.xlabel("x")
    plt.ylabel(f"x^{a}")
    plt.title(f"Plot of x^{a}")
    plt.show()

PlotPower(range(1, 11), 3)
```
